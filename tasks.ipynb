```python
from tasks import DelayGo, DelayAnti

import matplotlib.pyplot as plt
```

# Inspect Tasks

```python
# Select a task
task = DelayAnti()
```

Generate and plot a batch of task inputs and outputs

```python
smoothing_window = 10
batch_size = 3
noise = 0.05
period_duration = 25 

inputs, outputs = task.generate_batch(batch_size=batch_size, period_duration=period_duration, smoothing_window=smoothing_window, noise=noise)
task.plot_sequence(inputs, outputs)
```

Generate and plot all input/output task sequences

```python
inputs, outputs = task.generate_all_sequences(period_duration=period_duration)
task.plot_sequence(inputs, outputs)
```

```python
# Plot the mask 
mask = task.generate_mask(period_duration=period_duration, grace_frac = 0.2)

plt.imshow(mask, cmap='binary_r', aspect = '.1')
plt.xlabel('Mask')
plt.ylabel('Time Steps')
plt.colorbar()

# Remove x-axis labels
plt.xticks([])
plt.show()
```
