```python
from tasks import DelayGo, DelayAnti

import matplotlib.pyplot as plt
```

# Inspect Tasks

```python
# Select a task
task = DelayAnti(period_duration=50, grace_period = 10)
```

Generate and plot a batch of task inputs and outputs

```python
smoothing_window = 10
batch_size = 3
noise = 0.05

inputs, outputs = task.generate_batch(batch_size=batch_size, smoothing_window=smoothing_window, noise=noise)
task.plot_sequence(inputs, outputs)
```

Generate and plot all input/output task sequences

```python
inputs, outputs = task.generate_all_sequences()
task.plot_sequence(inputs, outputs)
```

```python
# Plot the mask 
mask = task.mask

plt.imshow(mask, cmap='binary_r', aspect = '.1')
plt.xlabel('Mask')
plt.ylabel('Time Steps')
plt.colorbar()

# Remove x-axis labels
plt.xticks([])
plt.show()
```
